```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("bhagavad-gita-in-english-source-file.pdf")
data = loader.load()

# Combine all pages into one big string
gita_text = "\n".join([doc.page_content for doc in data])
```

# STEP-3: Split Text into Verses

```python
import pickle

try:
    # Try loading saved verses
    with open("gita_verses.pkl", "rb") as f:
        verses = pickle.load(f)
    print("Verses loaded successfully!")
except FileNotFoundError:
    # If the file doesn't exist, split and clean verses and save them
    verses = gita_text.split('\n')
    verses = [v.strip() for v in verses if v.strip() != ""]
    with open("gita_verses.pkl", "wb") as f:
        pickle.dump(verses, f)
    print("Verses saved for the next time.")
```

#  Step-4: Create Embeddings & Store in FAISS
use below code only 1st time
Because no need to save again and again

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert verses to vectors
embeddings = model.encode(verses)

# Create FAISS index
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(np.array(embeddings))

# Save index
faiss.write_index(index, "gita_index.faiss")

print(" Embeddings created and FAISS index saved!")
```

```python
import pickle, faiss
from sentence_transformers import SentenceTransformer

# Load verses
with open("gita_verses.pkl", "rb") as f:
    verses = pickle.load(f)

# Load FAISS index
index = faiss.read_index("gita_index.faiss")

# Ask question
model = SentenceTransformer('all-MiniLM-L6-v2')
query = "what is dharma?"
query_vector = model.encode([query])

# Search top match
# D, I = index.search(query_vector, k=3)

# print("🕉️ Gita Answer:")
# print(verses[I[0][0]])

# Search top 3 matches
D, I = index.search(query_vector, k=3)

print("🕉️ Gita Answers:")
for i, idx in enumerate(I[0]):
    print(f"Rank {i+1}:")
    print(f"Verse: {verses[idx]}")
    print(f"Distance: {D[0][i]}")
    print("------------")

```

# Step-7: Functions for Answering

# A- Smart AI mode (RAG style)

```python
def answer_question(question):
    q_embed = model.encode([question])
    D, I = index.search(np.array(q_embed), 3)  # Top 3 matches
    matched_verses = [verses[i] for i in I[0]]
    context = "\n".join(matched_verses)

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a spiritual teacher answering using the Bhagavad Gita."},
            {"role": "user", "content": f"My question: {question}\n\nRelevant verses:\n{context}"}
        ]
    )
    return response['choices'][0]['message']['content']
```

# B- Random verse mode (prashanavali)

```python
import random

def explain_random_verse():
    verse = random.choice(verses)

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a spiritual teacher explaining Bhagavad Gita verses."},
            {"role": "user", "content": f"Explain this Gita verse in simple words:\n\n{verse}"}
        ]
    )
    return f"🕉️ Random Verse:\n\n{verse}\n\n📖 Explanation:\n{response['choices'][0]['message']['content']}"
```

# Step-8: Gradio UI with Mode Switch

```python
import gradio as gr

def chatbot_interface(question, mode):
    if mode == "Smart AI Answer":
        return answer_question(question)
    else:
        return explain_random_verse()

gr.Interface(
    fn=chatbot_interface,
    inputs=[
        gr.Textbox(label="Your Question"),
        gr.Radio(["Smart AI Answer", "Random Verse"], label="Choose Mode")
    ],
    outputs="text",
    title="🕉️ Bhagavad Gita Chatbot",
    description="Ask any life question or get a divine message from a random verse."
).launch(debug=True)
```
